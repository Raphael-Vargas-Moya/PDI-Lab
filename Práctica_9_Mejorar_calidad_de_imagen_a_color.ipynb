# **Glaucoma**

Primero cargamos la imagen y la convertimos a HSV. Luego separamos los canales para posteriormente ecualizar el canal del brillo, después combinamos los canales y finalmente convertimos de vuelta a RGB.

```python
# Cargar imagen oscura
img_bgr = cv2.imread('/content/glaucoma.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Convertir a HSV
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# Separar canales H, S y V
h, s, v = cv2.split(img_hsv)

# Ecualizar solo el canal de valor (brillo)
v_eq = cv2.equalizeHist(v)

# Combinar canales con el V ecualizado
img_hsv_eq = cv2.merge([h, s, v_eq])

# Convertir de nuevo a BGR y luego a RGB para visualización
img_eq_bgr = cv2.cvtColor(img_hsv_eq, cv2.COLOR_HSV2BGR)
img_eq_rgb = cv2.cvtColor(img_eq_bgr, cv2.COLOR_BGR2RGB)
```

Ocupamos Matplotlib para su visualización, quedando en este caso (y en los posteriores) una imagen considerablemente peor, debido a que este tipo de imágenes se toman con una iluminación alta, la cual es necesaria para atravesar los tejidos y apreciar las diferencias, por lo que al ecualizar dicho brillo, las diferencias se difuminan.

```python
# Mostrar imágenes original y mejorada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_eq_rgb)
plt.title('Imagen mejorada (Ecualización V)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
